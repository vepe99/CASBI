```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_parquet('/export/data/vgiusepp/data/full_dataframe/dataframe/dataframe.parquet')
```

```python
plt.hist(dataframe['star_log10mass'].unique())
```

```python
# Assuming your dataframe is named 'dataframe'
# Group by 'star_log10mass' and count the occurrences
dataframe['star_log10mass'] = 10**dataframe['star_log10mass']
mass_counts = dataframe.groupby('star_log10mass').size().reset_index(name='counts')

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(mass_counts['star_log10mass'], mass_counts['counts'], alpha=0.6)
plt.xlabel('star_log10mass')
plt.ylabel('Number of Rows')
plt.title('Scatter Plot of star_mass vs Number of Rows')
plt.grid(True)
plt.show()
```
